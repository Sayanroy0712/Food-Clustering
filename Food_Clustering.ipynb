```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/recipedata_bca.csv")
df.head()
```

```python
# @title Region

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Region').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Spiciness

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Spiciness').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Calories per Serving

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Calories per Serving').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Protein per Serving

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Protein per Serving').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Carbs per Serving

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Carbs per Serving').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Region vs Spiciness

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Spiciness'].value_counts()
    for x_label, grp in df.groupby('Region')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Region')
_ = plt.ylabel('Spiciness')
```
